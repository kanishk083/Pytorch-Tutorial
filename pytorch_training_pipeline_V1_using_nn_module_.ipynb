#Create Model

```python
import torch.nn as nn

class Model(nn.Module):
  def __init__(self,no_features):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(no_features,4),
        nn.ReLU(),
        nn.Linear(4,2),
        nn.ReLU(),
        nn.Linear(2,1),
        nn.Sigmoid()
    )

  def forward(self,features):
    output = self.network(features)
    return output

```

#Paramters

```python
learning_rate = 0.1
epochs = 25
```

```python
#Loss function
loss_fn = nn.BCELoss()
```

#Training Pipelines

```python
#model
model = Model(X_train_tensors.shape[1])

#optemizer
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)

#loop
for epoch in range(epochs):
  #forwaed pass
  y_pred=model(X_train_tensors)

  #loss cal
  loss = loss_fn(y_pred,y_train_tensors.view(-1,1))

  #clear gradients
  optimizer.zero_grad()

  #backward pass
  loss.backward()

  #paramters update
  optimizer.step()

  #print epochs
  print(f'epoch: {epoch + 1} loss: {loss}')
```

#Evaluation

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensors)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tensors).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
